```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import glob

!unzip -q engagement_data.zip

try:
    districts_df = pd.read_csv("districts_info.csv")
    products_df = pd.read_csv("products_info.csv")

    print("Archivos 'districts_info.csv' y 'products_info.csv' cargados.")
    print("\nInformación de Distritos (primeras filas):")
    display(districts_df.head())
    print("\nInformación de Productos (primeras filas):")
    display(products_df.head())

except FileNotFoundError:
    print("ERROR: Asegúrate de haber subido 'districts_info.csv' y 'products_info.csv'")

path_to_engagement_files = "engagement_data"

all_files = glob.glob(path_to_engagement_files + "/*.csv")

engagement_list = []
for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    district_id = int(filename.split('/')[-1].split('.')[0])
    df['district_id'] = district_id
    engagement_list.append(df)

engagement_df = pd.concat(engagement_list, axis=0, ignore_index=True)

print(f"\n¡Éxito! Se cargaron y combinaron {len(all_files)} archivos de 'engagement'.")
print("\nInformación de 'Engagement' (primeras filas):")
display(engagement_df.head())
```

```python
import pandas as pd
import numpy as np

engagement_df['time'] = pd.to_datetime(engagement_df['time'])

def convert_range_to_numeric(x):
    if isinstance(x, str):
        parts = x.strip('[]').split(', ')
        if len(parts) == 2:
            try:
                lower = float(parts[0])
                upper = float(parts[1])
                return (lower + upper) / 2
            except ValueError:
                return np.nan
    return x

districts_df['county_connections_ratio'] = districts_df['county_connections_ratio'].apply(convert_range_to_numeric)

districts_df['locale'] = districts_df['locale'].fillna('Unknown')

numeric_cols = ['pct_black/hispanic', 'pct_free/reduced', 'county_connections_ratio', 'pp_total_raw']

for col in numeric_cols:
    districts_df[col] = pd.to_numeric(districts_df[col], errors='coerce')

print("Valores nulos en 'districts_df' (antes de limpiar):")
print(districts_df.isnull().sum())

for col in numeric_cols:
    state_mean = districts_df.groupby('state')[col].transform('mean')
    districts_df[col] = districts_df[col].fillna(state_mean)

districts_df = districts_df.fillna(districts_df.mean(numeric_only=True))

print("\nValores nulos en 'districts_df' (después de limpiar):")
print(districts_df.isnull().sum())

full_data_df = pd.merge(engagement_df, districts_df, on='district_id')

full_data_df = pd.merge(full_data_df, products_df, left_on='lp_id', right_on='LP ID')

print("\n¡DataFrames unidos exitosamente!")
print("Columnas del DataFrame combinado (full_data_df):")
print(full_data_df.columns)

print("\nDimensiones del DataFrame final:")
display(full_data_df.shape)

print("\nPrimeras filas del DataFrame final:")
display(full_data_df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

print("--- Iniciando Análisis Exploratorio (EDA) ---")

full_data_df['month'] = full_data_df['time'].dt.month
full_data_df['year'] = full_data_df['time'].dt.year
full_data_df['month_year'] = full_data_df['time'].dt.to_period('M')

engagement_over_time = full_data_df.groupby('month_year')['pct_access'].mean().reset_index()
engagement_over_time['month_year'] = engagement_over_time['month_year'].astype(str)

plt.figure(figsize=(15, 6))
sns.lineplot(data=engagement_over_time, x='month_year', y='pct_access')
plt.title('Promedio de % de Acceso a Productos Digitales (2020)', fontsize=16)
plt.xlabel('Mes', fontsize=12)
plt.ylabel('Promedio de pct_access', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

print("\n--- Correlación entre variables demográficas y de conectividad ---")

demographic_cols = ['pct_black/hispanic', 'pct_free/reduced', 'county_connections_ratio', 'pp_total_raw']
engagement_metrics = ['pct_access', 'engagement_index']

correlation_df = full_data_df[demographic_cols + engagement_metrics].corr()

plt.figure(figsize=(10, 7))
sns.heatmap(correlation_df, annot=True, fmt=".2f", cmap="coolwarm", vmin=-1, vmax=1)
plt.title('Mapa de Calor de Correlación', fontsize=16)
plt.show()

print("\n--- Relación entre 'locale' (zona) y 'engagement_index' ---")

plt.figure(figsize=(12, 6))
sns.boxplot(data=full_data_df, x='locale', y='engagement_index', order=['Rural', 'Suburb', 'Town', 'City', 'Unknown'])
plt.title('Distribución del Engagement Index por Tipo de Zona (Locale)', fontsize=16)
plt.xlabel('Tipo de Zona (Locale)', fontsize=12)
plt.ylabel('Engagement Index', fontsize=12)
plt.ylim(0, full_data_df['engagement_index'].quantile(0.95)) # Cortamos valores atípicos
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

print("--- Iniciando Parte II: Titanic ---")

try:
    train_df = pd.read_csv("train.csv")
    test_df = pd.read_csv("test.csv")
    print("Archivos 'train.csv' y 'test.csv' cargados.")
except FileNotFoundError:
    print("ERROR: Asegúrate de haber subido 'train.csv' y 'test.csv' de Titanic.")

print("\nPrimeras filas de 'train_df':")
display(train_df.head())

X = train_df.drop("Survived", axis=1)
y = train_df["Survived"]

numerical_features = ['Age', 'Fare', 'SibSp', 'Parch']
categorical_features = ['Embarked', 'Sex', 'Pclass']

numerical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numerical_transformer, numerical_features),
        ('cat', categorical_transformer, categorical_features)
    ])

model_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', RandomForestClassifier(n_estimators=100, random_state=42))
])

print("\n--- Entrenando el modelo ---")
model_pipeline.fit(X, y)
print("¡Modelo entrenado exitosamente!")

y_pred_train = model_pipeline.predict(X)
train_accuracy = accuracy_score(y, y_pred_train)
print(f"\nPrecisión (Accuracy) en los datos de entrenamiento: {train_accuracy:.4f}")

print("\n--- Generando predicciones para 'test.csv' ---")
predictions = model_pipeline.predict(test_df)

submission_df = pd.DataFrame({
    "PassengerId": test_df["PassengerId"],
    "Survived": predictions
})

submission_df.to_csv("submission.csv", index=False)

print("\nArchivo 'submission.csv' creado exitosamente.")
print("Primeras filas de la sumisión:")
display(submission_df.head())
```
